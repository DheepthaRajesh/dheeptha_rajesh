Understand the F-statistics

Larger F-value: A larger F-value indicates a greater difference among the group means. It suggests that the variations between the groups are significant.
Smaller F-value: Conversely, a smaller F-value suggests that the group means are similar, and there may not be significant differences among them.

Examine the P-Value


P-Value < Significance Level (e.g., 0.05): If the p-value is less than your chosen significance level (often set at 0.05), it indicates that there are statistically significant differences among the groups. In other words, you have evidence to reject the null hypothesis, which assumes no significant differences.
P-Value ≥ Significance Level: If the p-value is equal to or greater than the chosen significance level, you do not have enough evidence to reject the null hypothesis. This suggests that any observed differences are likely due to random chance.

## 3. Data Modelling :

### Feedforward Neural Network :

```python
# import the necessary libraries :

import torch
import torch.optim as optim
import torch.nn as nn
import tqdm
import copy
```

```python
# Convert the X and y to 2D PyTorch tensors to be used in the FNN :

X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(np.array(y_train), dtype=torch.float32).reshape(-1, 1)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(np.array(y_test), dtype=torch.float32).reshape(-1, 1)
```

Define the model (and the model hyperparameters) :

```python
class FeedForwardNN(nn.Module):
  # input_size: Dimensionality of input feature vector.
  # num_hidden: The number of hidden layers to use.
  # hidden_dim: The size of each of the hidden layers.
  # dropout: The proportion of units to drop out after each layer.
  def __init__(self, input_size, num_hidden, hidden_dim, dropout):

    super(FeedForwardNN, self).__init__()

    assert num_hidden > 0

    self.hidden_layers = nn.ModuleList([])

    self.hidden_layers.append(nn.Linear(input_size, hidden_dim))


    for i in range(num_hidden - 1):
      self.hidden_layers.append(nn.Linear(hidden_dim, hidden_dim))

    # Set up the dropout layer.
    self.dropout = nn.Dropout(dropout)

    # Output projection :
    self.output_projection = nn.Linear(hidden_dim, 1) # Output size is 1 for regression problem

    # We use linear activation function which is commonly used for regression tasks where the output is a real number and cannot be squashed into a range


  # input is of shape (batch_size, input_size)
  def forward(self, x):

    for hidden_layer in self.hidden_layers:
      x = hidden_layer(x)
      x = self.dropout(x)

    # Output layer
    out = self.output_projection(x)

    return out
```

```python
# Model hyperparameters :

# Input_size - Dimensionality of the input feature vector - No. of features we are considering for the model :

input_size = X_train.shape[1]
print("Input size: ",input_size)

# Num_hidden - Number of hidden layers - 2 (trial and error method)
num_hidden = 2

# Hidden_dim - Size of each hidden layer - Approximately the average of input size and output size
hidden_dim = int((input_size + 1)/2)
print("Size of each hidden layer: ",hidden_dim)

# Dropout - Dropout rate
dropout = 0.2
```

```python
ffnn_model = FeedForwardNN(input_size, num_hidden,
                         hidden_dim, dropout)
print(ffnn_model)

parameters = ffnn_model.parameters()

print("Shapes of model parameters:")
print([x.size() for x in list(parameters)])
```

```python
# Set model optimizer as Adam - adaptive learning rate :

# Add weight_decay = 0.01 for L2 regularization :

ffnn_optimizer = optim.Adam(ffnn_model.parameters(), lr = 0.0001, weight_decay = 0.01)
```

```python
# Set loss function as Mean Squared Error (suitable for regression problems)

ffnn_loss_func = nn.MSELoss()
```

```python
# Set loss function as Mean Squared Error (suitable for regression problems)

ffnn_loss_func = nn.MSELoss()
```

```python
# Training parameters :

num_epochs = 100

batch_size = 10

start_batch = torch.arange(0, len(X_train), batch_size)
```

```python
# Store the best model after iterating through :

best_mse = np.inf
best_weights = None
list_mse = []
```

Train the model :

```python
# Training the model :

for epoch in range(num_epochs):
  ffnn_model.train()
  with tqdm.tqdm(start_batch, unit="batch", mininterval=0, disable=True) as progress_bar:
      progress_bar.set_description(f"Epoch {epoch}")
      for start in progress_bar:

          X_batch = X_train[start:start+batch_size]
          y_batch = y_train[start:start+batch_size]
          # Forward pass
          y_pred = ffnn_model(X_batch)
          loss = ffnn_loss_func(y_pred, y_batch)
          # Backward pass
          ffnn_optimizer.zero_grad()
          loss.backward()

          ffnn_optimizer.step()
          # print progress
          progress_bar.set_postfix(mse=float(loss))
  # evaluate mse at end of each epoch
  ffnn_model.eval()
  y_pred = ffnn_model(X_test)
  mse = ffnn_loss_func(y_pred, y_test)
  mse = float(mse)
  list_mse.append(mse)
  if mse < best_mse:
      best_mse = mse
      best_weights = copy.deepcopy(ffnn_model.state_dict())


# Reference : https://machinelearningmastery.com/building-a-regression-model-in-pytorch/
```

Evaluate the model based on MSE and RMSE metrics :

```python
# Get the best model
ffnn_model.load_state_dict(best_weights)

# Print best MSE and RMSE
print("Best MSE: %.2f" % best_mse)
print("Best RMSE: %.2f" % np.sqrt(best_mse))

# Plot the history of MSE
plt.plot(list_mse)
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.title('MSE vs Epochs')
plt.show()
```

```python
# To get range of target variable to analyse the MSE and RMSE values :

min_price = df_final['Price'].min()
max_price = df_final['Price'].max()

print("Minimum price:", min_price)
print("Maximum price:", max_price)
```

```python
# Test the model on 5 datapoints from the test dataset :

ffnn_model.eval()
with torch.no_grad():
    
    for i in range(5):
        X_sample = X_test_original.iloc[i:i+1]
        X_sample = scaler.transform(X_sample)
        X_sample = torch.tensor(X_sample, dtype=torch.float32)
        y_pred = ffnn_model(X_sample)
        print(f"Expected Price: {y_test[i].numpy()} - Predicted Price: {y_pred[0].numpy()}")
```

```python
# Calculate accuracy based on a threshold :

threshold = 0.02  # Define the threshold as ±2% of the actual price

correct_predictions = 0
total_samples = len(y_test)

ffnn_model.eval()
with torch.no_grad():
    for i in range(total_samples):
        X_sample = X_test_original.iloc[i:i+1]
        X_sample = scaler.transform(X_sample)
        X_sample = torch.tensor(X_sample, dtype=torch.float32)
        y_pred = ffnn_model(X_sample)

        expected_price = y_test[i].numpy()
        predicted_price = y_pred[0].numpy()

        if abs(expected_price - predicted_price) <= threshold * expected_price:
          correct_predictions += 1

accuracy = (correct_predictions / total_samples) * 100
print("Accuracy: ", accuracy, "%")
```
